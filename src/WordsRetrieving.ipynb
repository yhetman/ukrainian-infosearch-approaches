```python
import re
from string import digits
```

```python
book = open('../books/Intermezzo.txt', 'r')
content = book.read()
```

```python
raw_content = " ".join(content.split())
raw_content
```

```python
with open('parenthesis.txt', 'r') as f:
    parenthesis = f.read()
    
parenthesis = parenthesis.split(', ')
```

```python
drop_parenthesis = [i for i in raw_content.split(', ') if i.lower() not in parenthesis]
drop_parenthesis = ", ".join(drop_parenthesis).replace('-', ' ')
drop_parenthesis
```

```python
punc = '''!()-[]{};:"\,<>./?@#$%^&*_~'''

for ele in raw_content:
    if ele in punc:
        raw_content = raw_content.replace(ele, "")


table = str.maketrans('', '', digits)
raw_content = raw_content.translate(table)

words = raw_content.split(" ")
filtered_words = [i for i in words if i != '']
filtered_words
```
