## Reading files

```python
%%time
import re, os, json
from string import digits
import pandas as pd

docIDs = list(range(0, 18))
def delete_sings_digits(content):
    punc = '''!()-[]{};:"\,«»<>./?@#$%—…^&*_~'''
    for ele in content:
        if ele in punc:
            content = content.replace(ele, "")
    table = str.maketrans('', '', digits)
    content = content.translate(table)
    return content


def create_vocabulary(content):    
    raw_content = delete_sings_digits(content)
    vocab = [i.lower() for i in raw_content.split() if i != '']
    return vocab

def create_word_pairs(text):
    pairs = [text[i] + ' ' + text[i + 1] for i in range(len(text) - 1)]
    return pairs
    
def read_books(dirname='../books/'):
    books = os.listdir(dirname)
    vocabulary = list()
    pairs = [0] * len(books)
    dictionary = dict()
    for i, book in enumerate(books):
        with open(dirname + book, 'r', encoding='utf-8') as file:
            text = create_vocabulary(file.read())
            pairs[i] = create_word_pairs(text)
            dictionary[book.replace('.txt', '')] = text
            vocabulary += text
    return pairs, vocabulary, dictionary


pairs, vocabulary, dictionary = read_books()
print(f'Total number of words : {len(vocabulary)}')
print(f'Total number of tokens : {len(set(vocabulary))}')
```

```python
all_pairs = list()
for i in pairs:
    all_pairs += i
print(f'Total number of two-word pairs : {len(all_pairs)}')
print(f'Total number of unique two-word pairs : {len(set(all_pairs))}')
```

## Function to create Incidence Matrix

```python
%%time

def pair_incidence_matrix(pairs, all_pairs, dirname='../books/'):
    matrix = list()
    books = os.listdir(dirname)
    for i, words in enumerate(pairs):
        uni_pairs = list(set(words))
        column = dict()
        for word in uni_pairs: 
            column[word] = 1
        other = list(set(all_pairs) - set(uni_pairs))
        for word in other:
            column[word] = 0
        matrix.append(column)
    matrix = pd.DataFrame(matrix, index = books, columns = list(set(all_pairs)))
    return matrix.transpose()

matrix = pair_incidence_matrix(pairs, all_pairs)
```

## Function to create Inverted Indexes

```python
%%time
def pair_invert_index(all_pairs, pairs):
    uniq = list(set(all_pairs))
    voc = dict(zip(uniq, range(len(uniq))))
    freq = [0] * len(voc)
    docIDs = dict(zip(uniq, [-1] * len(uniq)))

    for j, words in enumerate(pairs):
        print('Processing of', j, "array of pairs started...")
        uni_words = list(set(pairs[j]))
        for word in uni_words:
            freq[voc[word]] += 1
            if docIDs[word] != -1:
                docIDs[word] += [j]
            else:
                docIDs[word] = [j]
    invert_indx = dict()
    for word in list(voc.keys()):
        invert_indx[word] = dict()
        invert_indx[word]['frequency'] = freq[voc[word]]
        invert_indx[word]['docIDs'] = docIDs[word]
    return invert_indx

pair_invert_indx = pair_invert_index(all_pairs, pairs)
```

```python
def AND(arr1, arr2):
    result = len(arr1) * [0]
    for i in range(len(arr1)):
        result[i] = 1 if arr1[i] == 1 and arr2[i] == 1 else 0
    return result
```

```python
def OR(arr1, arr2):
    result = len(arr1) * [0]
    for i in range(len(arr1)):
        result[i] = 1 if arr1[i] > 0 or arr2[i] > 0 else 0
    return result
```

```python
def NOT(word):
    return [abs(i - 1) for i in word]
```

```python
def AND_ind(word1, word2):
    return list(set(word1) & set(word2))
```

```python
def OR_ind(word1, word2):
    return list(set(word1) | set(word2))
```

```python
def NOT_ind(word, docIDs=docIDs):
    return [i for i in docIDs if i not in word]           
```
