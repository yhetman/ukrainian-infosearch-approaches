```python
import nltk as nltk
import re
```

```python
book = open('../books/Intermezzo.txt', 'r')
content = book.read()
```

```python
raw_content = content.replace('\n', " ").replace('\t', " ").replace('  ', " ").replace('                 ', " ").replace('           ', " ").replace('  ', " ")
raw_content
```

```python
with open('parenthesis.txt', 'r') as f:
    parenthesis = f.read()
    
parenthesis = parenthesis.split(', ')
parenthesis
```

```python

drop_parenthesis = [i for i in raw_content.split(', ') if i.lower() not in parenthesis]
drop_parenthesis = ", ".join(drop_parenthesis).replace('-', ' ')
drop_parenthesis
```

```python
punc = '''!()-[]{};:"\,<>./?@#$%^&*_~'''

for ele in raw_content:
    if ele in punc:
        raw_content = raw_content.replace(ele, "")


words = raw_content.split(" ")
filtered_words = [i for i in words if i != '']
filtered_words
```

```python
cyrillic_letters = r'\b[А-Я]\w*'
latin_letters = r'\b[A-Z]\w*'

capital_words =  re.findall(latin_letters, raw_content) + re.findall(cyrillic_letters, raw_content)
capital_words
```

```python
with open('pronouns.txt', 'r') as f:
    pronouns = f.read()
    
p = pronouns.replace('\n', ' ')
pronouns = p.split(', ')
# pronouns
```

```python
with open('particula.txt', 'r') as f:
    particula = f.read()
    
particula = particula.split(', ')
# particula
```
