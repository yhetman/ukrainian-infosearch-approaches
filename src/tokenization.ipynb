```python
import re
from string import digits
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
book = open('../books/Intermezzo.txt', 'r')
content = book.read()
raw_content = " ".join(content.split())

punc = '''!()-[]{};:"\,<>/?@#$%^&*_~'''

for ele in raw_content:
    if ele in punc:
        raw_content = raw_content.replace(ele, "")


table = str.maketrans('', '', digits)
raw_content = raw_content.translate(table)

words = raw_content.split(" ")
filtered_words = [i for i in words if i != '']
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words = 1500)
tokenizer.fit_on_texts(filtered_words)
word_index = tokenizer.word_index
word_index
```

```python
sequences = tokenizer.texts_to_sequences(raw_content.replace('...', '.').split('.'))
sequences[:2]
```

```python
padded = pad_sequences(sequences)
padded
```

```python
 model = tf.keras.Sequential([
     tf.keras.layers.Embedding(1840, 64),
     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences = True)),
     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
     tf.keras.layers.Dense(64, activation = 'relu'),
     tf.keras.layers.Dense(1, activation = 'sigmoid')    
 ])
model.summary()
```
